```python
# a. 
import numpy as np
A = np.array([[2, 1, 1],[1, 2, 1],[1, 1, 2]])

# Kim tra kh nghch
det_A = np.linalg.det(A)
print("nh thc ca A =", det_A)

A_inv = np.linalg.inv(A)
print("Ma trn nghch o ca A:\n", A_inv)
```

```python
# b. 
name = "Tran Nhat Thinh"
```

```python
# c.
# Chuyn tng k t trong tn sang m ASCII
ascii_codes = [ord(c) for c in name]
# B sung 0  s phn t chia ht cho 3
while len(ascii_codes) % 3 != 0:
    ascii_codes.append(0)
# Chuyn danh sch ASCII thnh ma trn (mi hng 3 phn t)
matrix_text = np.array(ascii_codes).reshape(-1, 3)
# M ha bng ma trn A
encoded = matrix_text @ A
print("Ma trn m ha:\n", encoded)
```

```python
# d. 
# Ma trn nghch o
A_inv = np.linalg.inv(A)

# Gii m
decoded = encoded @ A_inv
decoded = np.round(decoded).astype(int)

# Chuyn ASCII v chui
decoded_name = ""
for row in decoded:
    for num in row:
        if num != 0:
            decoded_name += chr(num)

print("Chui sau khi gii m:", decoded_name)
```
